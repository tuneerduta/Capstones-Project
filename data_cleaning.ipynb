# Modification of Columns

```python
df_new = df['Skills'].str.split(', ',expand=True)
```

```python
df_new = df_new.iloc[:,0:4]
```

```python
df_new.rename(columns = {0:'Skill_1',1:'Skill_2',2:'Skill_3',3:'Skill_4'}, inplace = True)
```

```python
for i in df_new.columns:
    df[i] = df_new[i]
```

```python
df.drop('Skills',inplace=True,axis=1)
```

```python
df_new = df['Locations'].str.split(', ',expand=True)
```

```python
df_new.drop(2,inplace=True,axis=1)
```

```python
df_new.rename(columns = {0:'Location',1:'Location_2'}, inplace = True)
```

```python
for i in df_new.columns:
    df[i] = df_new[i]
```

```python
df.drop('Locations',axis=1,inplace=True)
```

```python
df.drop('Skill_4',axis=1,inplace=True)
```

```python
df.drop('Location_2',axis=1,inplace=True)
```

```python
df['Job_id'] = df['Link'].str.extract('(\d+)')
df.drop(['Link'],axis=1,inplace=True)
```

```python
df.rename(columns = {'Established Year':'Establishment_Year'}, inplace = True)
df.rename(columns = {'Employees Count':'Employees_Count_Category'}, inplace = True)
```

# Dealing with the Noise Values

```python
for i in df.columns:
    print(i,df[i].sort_values().unique(),'\n',sep='\n')
```

```python
df['Skill_2'] = df['Skill_2'].replace('#','NA')
```

```python
df['Designation'] = df['Designation'].str.removesuffix('(Drive)')
```

# Category Values Replacement

```python
df['Employees_Count_Category'].sort_values().unique()
```

```python
df['Employees_Count'] = df['Employees_Count_Category']
```

```python
# more than 1000 employees - Very High
# 500 - 1000 employees - High
# 200 - 500 employees - Semi High
# 50 - 200 employees - Medium
# 10 - 50 employees - Low
# 0 - 10 employees - Very Low
```

```python
df['Employees_Count_Category'] = df['Employees_Count_Category'].replace('0 - 10 employees','Very Low')
df['Employees_Count_Category'] = df['Employees_Count_Category'].replace('10 - 50 employees','Low')
df['Employees_Count_Category'] = df['Employees_Count_Category'].replace('50 - 200 employees','Medium')
df['Employees_Count_Category'] = df['Employees_Count_Category'].replace('200 - 500 employees','Semi High')
df['Employees_Count_Category'] = df['Employees_Count_Category'].replace('500 - 1000 employees','High')
df['Employees_Count_Category'] = df['Employees_Count_Category'].replace('More than 1000 employees','Very High')
```

# Dealing with the null values

```python
df['Skill_2'] = df['Skill_2'].fillna('NA')
df['Skill_3'] = df['Skill_3'].fillna('NA')
```
